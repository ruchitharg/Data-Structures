```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

import numpy as np
import matplotlib.pyplot as plt

# Generate dataset with categories and numerical values
np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = {
    'Category A': np.random.normal(20, 5, 100),
    'Category B': np.random.normal(30, 10, 100),
    'Category C': np.random.normal(25, 7, 100),
    'Category D': np.random.normal(35, 15, 100)
}

# Prepare data for box plot
data = [values[category] for category in categories]

# Create box plot
plt.boxplot(data, labels=categories)
plt.title('Distribution of Numerical Values Across Categories')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility
n_samples = 100
feature1 = np.random.rand(n_samples)
feature2 = 2 * feature1 + np.random.randn(n_samples) * 0.1  # Strongly correlated with feature1
feature3 = feature1 + feature2 + np.random.randn(n_samples) * 0.1  # Correlated with both feature1 and feature2
feature4 = np.random.rand(n_samples)  # Not correlated

# Create a DataFrame
data = pd.DataFrame({
    'Feature1': feature1,
    'Feature2': feature2,
    'Feature3': feature3,
    'Feature4': feature4
})

# Compute the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Generate dataset for sales data
np.random.seed(20)
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data_sales)

# Pivot the data to create a matrix format suitable for heatmap
pivot_table = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='sum', fill_value=0)

# Create heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap of Sales Data Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

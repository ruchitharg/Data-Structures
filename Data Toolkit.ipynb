```python
# 3. Explain the following terms:
#-The difference in np.array, np.asarray and np.asanyarrayX

'''
np.array:

Purpose: Used to create a new array from an object, typically a list, tuple, or another array.
Behavior: Always creates a new array, copying the data by default.
Parameters:  object: The input data to be converted to an array.
             dtype: The desired data type of the array.
             copy: A boolean indicating whether to copy the data (default is True).
'''                     
import numpy as np
data = [1, 2, 3]
arr = np.array(data)

'''
np.asarray:

Purpose: Converts the input to an array, but does not copy the data if the input is already an array with the same dtype.
Behavior: If the input is already an array of the same dtype, it returns the input array itself without copying. Otherwise, it creates a new array.
Parameters:  a: The input data to be converted to an array.
             dtype: The desired data type of the array.
'''    

data = [1, 2, 3]
arr = np.asarray(data)

'''
np.asanyarray:
Purpose: Similar to np.asarray, but it passes through subclasses of ndarray. This is useful when working with custom array types that are subclasses of ndarray.
Behavior: If the input is an instance of ndarray or a subclass, it returns the input itself. Otherwise, it converts the input to an array.
Parameters:  a: The input data to be converted to an array.
             dtype: The desired data type of the array.
'''
data = [1, 2, 3]
arr = np.asanyarray(data)

#- The difference between Deep copy and shallow copyX

'''
Shallow Copy:
Definition: A shallow copy creates a new object, but does not create copies of the objects contained within the original object. Instead, it inserts references to the original objects.
Behavior: If the original object contains references to other objects, the shallow copy will reference the same objects.
Usage: Can be created using the copy module in Python, or using the copy method for lists.
'''
import copy
original = [1, 2, [3, 4]]
shallow = copy.copy(original)

'''
Deep Copy:
Definition: A deep copy creates a new object and recursively copies all objects found in the original object, effectively duplicating everything.
Behavior: Changes made to the deep copy do not affect the original object and vice versa.
Usage: Can be created using the copy module in Python.
'''
import copy
original = [1, 2, [3, 4]]
deep = copy.deepcopy(original)
```

```python
#14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' andright y-axis for 'Humidity').

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate date range
date_range = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

# Generate random data for Temperature and Humidity
np.random.seed(0)
temperature_data = np.random.uniform(low=-10, high=35, size=len(date_range))
humidity_data = np.random.uniform(low=10, high=90, size=len(date_range))

# Create a DataFrame
data = {
    'Date': date_range,
    'Temperature': temperature_data,
    'Humidity': humidity_data
}
df = pd.DataFrame(data)

# Plot Temperature and Humidity on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Show plot
fig.tight_layout()
plt.title('Temperature and Humidity over Time')
plt.show()

# Display the first few rows of the DataFrame to verify
print(df.head())
```

```python
#15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
#a) Plot a histogram of the data with 30 bins.

import numpy as np
import matplotlib.pyplot as plt

# Generate a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#b) Overlay a line plot representing the normal distribution's probability density function (PDF).

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Generate Data
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Plot Histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Step 3: Overlay PDF Line Plot
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, p, 'k', linewidth=2)

title = "Histogram and PDF of Normal Distribution"
plt.title(title)
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
#d) Set the title of the plot as 'Histogram with PDF Overlay'.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Create the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Calculate the PDF and overlay it
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Add the title
plt.title('Histogram with PDF Overlay')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
quadrants = ['I' if (xi > 0 and yi > 0) else 
             'II' if (xi < 0 and yi > 0) else 
             'III' if (xi < 0 and yi < 0) else 
             'IV' for xi, yi in zip(x, y)]

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'quadrant': quadrants})

# Create a scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=data, x='x', y='y', hue='quadrant', palette='viridis')
scatter_plot.set_title('Quadrant-wise Scatter Plot')
scatter_plot.set_xlabel('X Axis')
scatter_plot.set_ylabel('Y Axis')
plt.legend(title='Quadrant')
plt.show()
```

```python
# 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'SineWave Function'.

import numpy as np
import pandas as pd
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

# Create the data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Output to a notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(title="SineWave Function", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Add grid lines
p.xgrid.visible = True
p.ygrid.visible = True

# Show the results
show(p)
```
